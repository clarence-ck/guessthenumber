```python
import random 

import itertools

print('Guess a correct number between 1 to 100 to win!')

value = 88 #random.randint(1, 100)

count = 0

numplayer = 0

Hostname =  input('Enter Hostname, or \'q\' to quit.')
  
if Hostname == "q" or Hostname.strip() == '' :
  print('No name registered')

else:
  print("Welcome Host {} to 'Guess the right number'".format(Hostname))

while numplayer != int:

  numplayer = input('Enter the number of players.')
  
  if numplayer.isnumeric():
    numplayer = int(numplayer)
    break
  
  else:
    print('Integer numbers only, please!')
 

playerindex = range(1,(numplayer+1))

Playerindex = []

for each in playerindex:
  x = str(each)
  Playerindex.append(x)

Player = []

for every in Playerindex:
  y = 'Player' + every
  Player.append(y)
  

Names=[]

for item in Player:
  item = input (f'Enter {item} name, or \'q\' to quit.')
  Names.append(item)
  if item == "q" or item.strip() == '' :
    print('No name registered')
  else:
    print("Welcome {} to 'Guess the right number'".format(item))

everyguess = 0

upperbound = 100
  
lowerbound = 1

nextname = 0

guessrange = range(lowerbound, upperbound+1)

while everyguess!= value:

  count+=1

  if len(guessrange)==1 and guessrange[0] == value :
    break
  
  list_name_cycle = itertools.cycle(Names)

  next(list_name_cycle)
  
  for eachname in Names:

    nextname = next(list_name_cycle)

    if everyguess == value:
      break
      
    for everyguess in eachname:
      
      if len(guessrange)==1 and guessrange[0] == value:
        break 
      
      everyguess = input(f'{eachname}, enter guess #{count}: ')
      
      modguess = everyguess.strip('-')
      
      if modguess.isnumeric():
        everyguess = int(everyguess)
     
      else:
        print('Integer numbers only, please!')
        continue
      
      if (lowerbound<=int(everyguess)<=upperbound):
        print('{} guessed number {}'.format(eachname,everyguess))
    
      else:
        print('{} guessed number {}, not a number within 1 to 100, try again'.format(eachname,everyguess))
        continue
      
      if everyguess > value:
        upperbound = range(lowerbound,(everyguess+1))[-1]
        guessrange = range(lowerbound,upperbound)
        upperbound -= 1
        
        if len(guessrange) != 1:
          print("{}'s guess {} is too high, try again!".format(eachname,everyguess))
          print(f'The lowerbound is {guessrange[0]} and upperbound is {guessrange[-1]}')
          print(list(guessrange))
          break
          
        elif len(guessrange) == 1:
          print(f'After {count} counts of guessing!, next player after {eachname}, {nextname} you are the joker! The jackpot is {value}')
          break
      
      elif everyguess < value:
        lowerbound = range(everyguess,upperbound)[0]
        guessrange = range((lowerbound+1),(upperbound+1))
        lowerbound += 1

        if len(guessrange) != 1 :
          print("{}'s guess {} is too low, try again!".format(eachname,everyguess))
          print(f'The lowerbound is {guessrange[0]} and upperbound is {guessrange[-1]}')
          print(list(guessrange))
          break

        elif len(guessrange) == 1:
          print(f'After {count} counts of guessing!, next player after {eachname}, {nextname} you are the joker! The jackpot is {value}')
          break

      elif everyguess == value:
        print(f'You guessed it in {count} tries!')
        break
      
       
    

      
      
     
```
