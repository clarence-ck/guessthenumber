```python
import random 

print('Guess a correct number between 1 to 100 to win!')

value = random.randint(1, 100)

count = 0

guess1 = 0

guess2 = 0

upperbound = 100

lowerbound = 1

Player1 = input('Enter Player 1 name, or \'q\' to quit.')
  
if Player1 == "q" or Player1.strip() == '' :
  print('No name registered')

else:
  print("Welcome {} to 'Guess the right number'".format(Player1))

Player2 = input('Enter Player 2 name, or \'q\' to quit.')

if Player2 == "q" or Player2.strip() == '' :
  print('No name registered')

else:
  print("Welcome {} to 'Guess the right number'".format(Player2))

while guess1 != value or guess2 != value:
  
  count +=1
  guess1 = input(f'{Player1}, enter guess #{count}: ')
  modguess1=guess1.strip('-')
  

  if modguess1.isnumeric():
    guess1 = int(guess1)
  
  else:
    print('Integer numbers only, please!')
    continue

  if (1<=int(guess1)<=100):
    print('{} guessed number {}'.format(Player1,guess1))
    
  else:
    print('{} guessed number {}, not a number within 1 to 100, try again'.format(Player1,guess1))
    continue
    
  if guess1 > value:
    upperbound = range(lowerbound,(guess1+1))[-1]
    guessrange = range(lowerbound,upperbound)
    upperbound -= 1
    print("{}'s guess {} is too high, try again!".format(Player1,guess1))
    print(f'The lowerbound is {guessrange[0]} and upperbound is {guessrange[-1]}')
    print(list(guessrange))
    if len(guessrange)==1:
      print(f'After {count} counts of guessing!, {Player2} you are the joker! The jackpot is {value}')
      break
  
  elif guess1 < value:
    lowerbound = range(guess1,upperbound)[0]
    guessrange = range((lowerbound+1),(upperbound+1))
    lowerbound += 1
    print("{}'s guess {} is too low, try again!".format(Player1,guess1))
    print(f'The lowerbound is {guessrange[0]} and upperbound is {guessrange[-1]}')
    print(list(guessrange))
    if len(guessrange)==1:
      print(f'After {count} counts of guessing!, {Player2} you are the joker! The jackpot is {value}')
      break

  elif guess1 == value:
    print(f'You guessed it in {count} tries!')
    break

  guess2 = input(f'{Player2}, enter guess #{count}: ')
  modguess2=guess2.strip('-')
  
  if modguess2.isnumeric():
    guess2 = int(guess2)
  
  else:
    print('Integer numbers only, please!')
    continue

  if (1<=int(guess2)<=100):
    print('{} guessed number {}'.format(Player2,guess2))
    
  else:
    print('{} guessed number {}, not a number within 1 to 100, try again'.format(Player2,guess2))
    continue
 
  if guess2 > value:
    upperbound = range(lowerbound,(guess2+1))[-1]
    guessrange = range(lowerbound,upperbound)
    upperbound -= 1
    print("{}'s guess {} is too high, try again!".format(Player2,guess2))
    print(f'The lowerbound is {guessrange[0]} and upperbound is {guessrange[-1]}')
    print(list(guessrange))
    if len(guessrange)==1:
      print(f'After {count} counts of guessing!, {Player1} you are the joker! The jackpot is {value}')
      break
  
  elif guess2 < value:
    lowerbound = range(guess2,upperbound)[0]
    guessrange = range((lowerbound+1),(upperbound+1))
    lowerbound += 1
    print("{}'s guess {} is too low, try again!".format(Player2,guess2))
    print(f'The lowerbound is {guessrange[0]} and upperbound is {guessrange[-1]}')
    print(list(guessrange))
    if len(guessrange)==1:
      print(f'After {count} counts of guessing!, {Player1} you are the joker! The jackpot is {value}')
      break
  
  elif guess2 == value:
    print(f'You guessed it in {count} tries!')
    break
```
